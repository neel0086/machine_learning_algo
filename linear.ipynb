```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from IPython.display import display
import pandas as pd
from sklearn.metrics import mean_squared_error
```

```python
data = datasets.load_diabetes()
df = pd.DataFrame(data.data, columns=data.feature_names)
pd.set_option("display.max_rows",None)
pd.set_option('display.memory_usage', True)
pd.set_option('display.expand_frame_repr', False)
print(df)

```

```python
diabetes_X = data.data[:,np.newaxis,2]
diabetes_train_X = diabetes_X
diabetes_test_X = diabetes_X

diabetes_Y = data.target
diabetes_train_Y = diabetes_Y
diabetes_test_Y = diabetes_Y


```

```python
model = linear_model.LinearRegression()
model.fit(diabetes_train_X,diabetes_train_Y)

diabetes_predict_Y = model.predict(diabetes_test_X)

print("Mean Squared error : " , mean_squared_error(diabetes_test_Y,diabetes_predict_Y))
print("Coeffiecent: ",model.coef_)
print("Intercept: ",model.intercept_)


```

```python
# Scatter Plot
plt.scatter(diabetes_test_X,diabetes_test_Y)
plt.plot(diabetes_test_X,diabetes_predict_Y)
plt.show()
```
