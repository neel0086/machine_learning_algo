```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = [1, 2, 3, 4]
Y = [3, 5, 7, 9]
w0 = np.random.randn()
w1=np.random.randn()
m=len(X)

alpha = 0.01
tolerance=10**(-6)
prev_cost = float('inf')

# y= w0+(x1)(w1)
# cost function 
# J(w) = (1/m)
```

```python
X=np.array(X)
Y=np.array(Y)
```

```python
#Training the model
# Gradient descent method
```

```python
for i in range(10000000):
    y_pred = w0 + w1*X
    error=y_pred-Y

    
    cost=(1/2*m)*np.sum(error**2)
    if abs(cost-prev_cost)<tolerance:
        break

    gradient_w0 = (1/m)* np.sum(error)
    gradient_w1 = (1/m)* np.sum(error*X)

    w0=w0-alpha*gradient_w0
    w1=w1-alpha*gradient_w1
    prev_cost=cost

regression_line=w0+w1*X
print(w0,w1)
plt.scatter(X,Y)
plt.plot(X,regression_line,label="Regression line ",color='red')
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
    
    
```
